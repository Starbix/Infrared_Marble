```python
import datetime
import colorcet as cc
import contextily as cx
from matplotlib.axes import Axes
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt


def plot(ax: Axes, date: str):
    dataset["Gap_Filled_DNB_BRDF-Corrected_NTL"].sel(time=date).plot.pcolormesh(
        ax=ax,
        cmap=cc.cm.bmy,
        robust=True,
    )
    assert gdf.crs is not None
    cx.add_basemap(ax, crs=gdf.crs.to_string())

    ax.text(
        0,
        -0.1,
        "Source: NASA Black Marble VNP46A2",
        ha="left",
        va="center",
        transform=ax.transAxes,
        fontsize=10,
        color="black",
        weight="normal",
    )
    ax.set_title(f"Myanmar: NTL Radiance on {date}", fontsize=20)


fig, ax = plt.subplots(1, 2, figsize=(16, 14))

plot(ax[0], "2018-09-18")
plot(ax[1], "2019-03-27")
```
